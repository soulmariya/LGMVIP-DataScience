**SPLITTING INTO TRAINING DATA AND TESTING DATA**

```python
X=iris.data[:,:]
Y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,random_state=1,test_size=0.1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=1) 
tree = DecisionTreeClassifier()
tree.fit(X_train,Y_train)
print("Training is done successfully.")
```

**PREDICTION ON TEST DATA**

```python
Y_pred = tree.predict(X_test)
print(Y_pred)
```

**ACCURACY OF THE MODEL**

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print("The Accuracy of our Model is: {}".format(accuracy_score(Y_test,Y_pred)))
```

The Decision Tree Model has a 100% accuracy.Therefore this model is good.

**PREDICTION USING NEW DATA**

```python
new_data =[[3.3, 4.0, 4.1, 2.3]]
Y_pred =tree.predict(new_data)
print(Y_pred)
```

**MODEL VISUALIZATION**

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
plt.figure(figsize  = (18,18))
tree.plot_tree(clf,filled = True,rounded = True,proportion = True,node_ids = True)
plt.show()
```
