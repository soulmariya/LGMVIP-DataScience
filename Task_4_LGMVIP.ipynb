### **IMPORTING REQUIRED LIBRARIES**

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow
```

### **READING THE IMAGE**

```python
image=cv2.imread("dog.jpg")
image
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### **CONVERTING TO GRAY SCALE**

```python
gray_img=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_img
```

```python
plt.imshow(gray_img)
```

### **INVERTING THE NEW GRAY SCALE IMAGE**

```python
invert=cv2.bitwise_not(gray_img)
invert
```

```python
plt.imshow(invert)
```

### **BLURRING THE IMAGE**

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
blur
plt.imshow(blur)
```

### **INVERTING THE BLURRED IMAGE**

```python
invertedblur=cv2.bitwise_not(blur)
invertedblur
```

```python
plt.imshow(invertedblur)
```

### **CONVERTING TO PENCIL SKETCH**

```python
pencil_sketch = cv2.divide(gray_img, invertedblur, scale=256.0)
plt.imshow(pencil_sketch)
cv2.waitKey(0)
```

### **COMPARISON BETWEEN ORIGINAL IMAGE AND PENCIL SKETCH**

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```

```python
cv2.imwrite("pencil_sketch.png",pencil_sketch)
```

```python
cv2.waitKey(0)
```
